# Top 5 most Medals won Countries, order the result by number of medals in descending order

```python
#Grouping by country
Medals_won_by_country=Medals_won_df.groupby(['reg'])['Medal'].count()

#sorting in descending order
Medals_won_by_country_desc=Medals_won_by_country.sort_values(ascending = False)

#Bring first 5 rows
Top_5_most_Medals_won_Countries=Medals_won_by_country_desc.head(5)

#reset index
Top_5_most_Medals_won_Countries=Top_5_most_Medals_won_Countries.reset_index()
Top_5_most_Medals_won_Countries
```

```python
#calculating percentages of Medals won
Top_5_most_Medals_won_Countries['percentage']=round(Top_5_most_Medals_won_Countries['Medal']/Top_5_most_Medals_won_Countries['Medal'].sum(),3)*100
Top_5_most_Medals_won_Countries
```

```python
# Doughnut Chart
import matplotlib.pyplot as plt
import mplcyberpunk
plt.style.use('cyberpunk')
 
top5_countries_won_medals=list(Top_5_most_Medals_won_Countries['reg'])
top5_countries_won_medals_count=list(Top_5_most_Medals_won_Countries['Medal'])

# colors
colors = ['#FF0000', '#0000FF', '#FFFF00',
          '#ADFF2F', '#FFA500']

# explosion
explode = (0.05, 0.05, 0.05, 0.05, 0.05)
 
# Pie Chart
plt.pie(top5_countries_won_medals_count, colors=colors, labels=top5_countries_won_medals,
        autopct='%1.1f%%', pctdistance=0.85,
        explode=explode)
 
# draw circle
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
 
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
 
# Adding Title of chart
plt.title('Top 5 most successful countries in olympics',color = 'white')
 
# Displaying Chart
plt.show()
```

```python
# Here I want to query using SQL in Jupiter notebook quickly.

# Here I am creating Olympic database.
import sqlite3 as sl
conn = sl.connect('Olympic.db')

print ("Opened database successfully")
```

```python
#Add a new table to your database called OlympicData using the data in the dataframe.
OlympicHistoryData = df_country.to_sql('OlympicHistoryData',conn, if_exists = 'replace')
```

```python
# Here I am checking the OlympicData created or not by displaying data.
pd.read_sql('select * from OlympicHistoryData limit 4', conn)
```

```python
#How many olympics games have been held?
Total_olympics_games_held=pd.read_sql('''select count(distinct Games) as Total_olympics_games_held 
                                               from OlympicHistoryData ''', conn)
Total_olympics_games_held
```

```python
#List down all Olympics games held so far. Order the result by year.
pd.read_sql('select distinct Games from OlympicHistoryData order by Year  ', conn)
```

```python
#Mention the total no of nations who participated in each olympics game?. Order the results by games
pd.read_sql('''select reg,count(distinct Games) as nations_participated  
               from OlympicHistoryData 
               group by reg 
               order by nations_participated desc''',conn)
```

```python
#Which nation has participated in all of the olympic games? and order the output by first column which is fetched
pd.read_sql('''select reg,count(distinct Games) as nations_participated  
               from OlympicHistoryData 
               group by reg 
               order by nations_participated desc limit 2''',conn)
```

```python
#How many unique athletes have won a gold medal in the Olympics?
no_of_athletes_won_gold_medal=pd.read_sql('''select count(distinct Name) as Athletes,Medal 
                                       from OlympicHistoryData 
                                       where Medal like "%Gold%" ''',conn)
no_of_athletes_won_gold_medal
```

```python
#Which Sports were just played only once in the olympics? 
Sports_played_once=pd.read_sql('''select Sport,Games ,count(distinct Games) as no_of_games
                                  from OlympicHistoryData 
                                  group by sport
                                  having count(distinct Games)=1
                                  order by Sport''', conn)
Sports_played_once
```

```python
#Fetch the total no of sports played in each olympic games. Order by no of sports by descending.

Sports=pd.read_sql('''select Games ,count(distinct Sport) as no_of_sports
                       from OlympicHistoryData 
                       group by Games
                       order by no_of_sports desc''', conn)
Sports
```

```python
#Fetch oldest athletes to win a gold medal
Top5_oldest_athletes=pd.read_sql('''select Age , 
                               Name ,Medal from OlympicHistoryData 
                               where Medal like '%Gold%' 
                               order by Age desc limit 5''',conn)
Top5_oldest_athletes
```

```python
#Top 5 athletes who have won the most gold medals. Order the results by gold medals in descending.
Top_5_athletes_who_have_won_the_most_gold_medals=pd.read_sql('''select Name,count(Medal) as Gold_Medals_Won,Medal 
                                                                from OlympicHistoryData 
                                                                group by Name,Medal
                                                                having Medal like '%Gold%' 
                                                                order by Gold_Medals_Won desc limit 5
                                                                ''',conn)
Top_5_athletes_who_have_won_the_most_gold_medals
```

```python
#Top 5 athletes who have won the most medals (gold/silver/bronze). Order the results by medals in descending.
pd.read_sql('''select Name,sum(Medals_Won) as No_Medals_won from(select Name,count(Medal) as Medals_Won,Medal 
               from OlympicHistoryData 
               group by Name,Medal
               having Medal like '%Gold%' or Medal like '%silver%' or Medal like '%bronze%'
               ) as Medals_count_by_Medal group by Name order by No_Medals_won desc limit 5
               
               ''',conn)
```

```python
#Top 5 most successful countries in olympics. Success is defined by no of medals won.
pd.read_sql('''select reg,sum(Medals_won) as No_of_Medals_won from (select reg,count(Medal) as Medals_won,Medal 
               from OlympicHistoryData 
               group by reg,Medal
               having Medal like '%Gold%' or Medal like '%Silver%' or Medal like '%Bronze%'
               ) as medals_count group by reg order by sum(Medals_won) desc limit 5
             ''',conn)
```

```python
#In which Sport/event, India has won highest medals.
pd.read_sql('''select reg,count(Medal) as Medals_won,Medal,Sport,Event 
               from OlympicHistoryData 
               group by reg,Medal,Event
               having (Medal like '%Gold%' or Medal like '%Silver%' or Medal like '%Bronze%') and reg='USA'
               order by count(Medal) desc limit 1
             ''',conn)
```

```python
#Break down all olympic games where india won medal for Hockey and how many medals in each olympic games and 
#order the result by no of medals in descending.

pd.read_sql('''select reg,count(Medal) as Medals_won,Medal,Sport,Event,Games 
               from OlympicHistoryData 
               group by reg,Medal,Event
               having (Medal like '%Gold%' or Medal like '%Silver%' or Medal like '%Bronze%') and reg='USA' and Sport='Athletics'
               order by count(Medal) desc limit 5
             ''',conn)
```
